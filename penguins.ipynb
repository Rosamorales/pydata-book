```python
from google.colab import files
from pyspark.sql.functions import col

uploaded = files.upload() # Selecciona el archivo que quieres cargar
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
df = spark.read.csv("penguins.csv", header=True, inferSchema=True) # Lee el archivo con pyspark
df.show() # Muestra el contenido del dataframe
```

```python
df_filtrado = df.select(col("rowid"), col("flipper_length_mm"), col("body_mass_g"), col("sex"))
```

```python
df_filtrado = df_filtrado.na.drop()

# Mostrar DataFrame sin valores nulos
df_filtrado.show()
```

```python
df_filtrado.show()
```

```python
from google.colab import files

uploaded = files.upload() # Selecciona el archivo que quieres cargar
df_penguins = spark.read.csv("penguins2.csv", header=True, inferSchema=True) # Lee el archivo con pyspark
df_penguins.show() # Muestra el contenido del dataframe
```

```python
df_penguins_filtrado = df_penguins.select(col("studyName"), col("Flipper Length (mm)"), col("Body Mass (g)"), col("Sex"))
```

```python
df_penguins_filtrado.show()
```

```python
df_penguins_filtrado = df_penguins_filtrado.na.drop()

# Mostrar DataFrame sin valores nulos
df_penguins_filtrado.show()
```

```python
# unir los dos conjuntos de datos en una sola tabla
df_unidos = df_filtrado.join(df_penguins_filtrado, 
                     (col("flipper_length_mm") == col("Flipper Length (mm)")) & 
                     (col("Body Mass (g)") == col("body_mass_g")), "inner")

# mostrar los resultados
df_unidos.show()
```

```python
df_unidos_outer = df_filtrado.join(df_penguins_filtrado, 
                     (col("flipper_length_mm") == col("Flipper Length (mm)")) & 
                     (col("Body Mass (g)") == col("body_mass_g")), how='full_outer')

# mostrar los resultados
df_unidos_outer.show()
```
